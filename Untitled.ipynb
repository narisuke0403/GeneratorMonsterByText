```python
import time
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://yakkun.com/data/zukan_ex5.htm"
```

```python
html = requests.get(url)
```

```python
soup = BeautifulSoup(html.content, "lxml")
```

```python
soup
```

```python
pokemon_name_tag = soup.find_all("h3")
```

```python
pokemon_name = []
for i in pokemon_name_tag:
    pokemon_name.append(i.text)
```

```python
pokemon_explain_tag = soup.select("p.flat")
```

```python
pokemon_explain = []
for i in pokemon_explain_tag:
    pokemon_explain.append(i.text)
pokemon_explain
```

```python
url = "https://yakkun.com/data/zukan/list.htm"
```

```python
html = requests.get(url)
```

```python
soup = BeautifulSoup(html.content, "lxml")
```

```python
soup
```

```python
all_pokemon_list_tag = soup.select("div.table")
```

```python
hitpoint = []
attack = []
defence = []
sp_attack = []
sp_defence = []
speed = []


for i in all_pokemon_list_tag[1:]:
    for t in i.find_all("tr"):
        for k, u in enumerate(t.find_all("td")):
            if k == 3:
                hitpoint.append(u.text)
            elif k == 4:
                attack.append(u.text)
            elif k == 5:
                defence.append(u.text)
            elif k  == 6:
                sp_attack.append(u.text)
            elif k == 7:
                sp_defence.append(u.text)
            elif k == 8:
                speed.append(u.text)
```

```python
pokemon_list = []
for i in range(len(pokemon_name)):
    pokemon = [pokemon_name[i], pokemon_explain[i], hitpoint[i], attack[i], defence[i], sp_attack[i], sp_defence[i], speed[i]]
    pokemon_list.append(pokemon)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(pokemon_list, columns=["name", "explain", "hitpoint", "attack", "defence", "sp_attack","sp_defence", "speed"])
```

```python
data
```

```python
data.to_csv("pokemon_data_v1.csv", index=False)
```
